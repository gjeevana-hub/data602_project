*    **Variable Identification**:
     
     
     1.   **Dependent Variable** (The "Effect" we are measuring): A dependent variable is the main outcome or response that is being measured and analyzed in a study. It is the variable that "depends on" or is hypothesized to be influenced by changes in the independent variable(s).
     This is the primary outcome we are trying to understand and predict.

          *Quality of Sleep*: This is the outcome we are trying to explain or predict. It is a continuous numerical variable representing the individual's self-reported quality of sleep. Higher values indicate better sleep quality, while lower values suggest poorer sleep quality. The study aims to understand which other variables, such as stress level, sleep duration, and lifestyle factors, are associated with the quality of an individual's sleep.


     2.  **Independent Variables** (The "Causes" or predictors we are testing): These are the variables we hypothesize might have a direct effect on the presence of a sleep disorder.

          *Sleep Duration (Continuous)* : We can test if the number of hours slept influences quality of sleep.

          *Sleep Disorder (Categorical)* : We can test if having a diagnosed sleep disorder (e.g., "Insomnia" or "Sleep Apnea") affects overall sleep quality.

          *Physical Activity Level (Continuous)* : We can test if the daily activity level (in minutes, for example) is associated with better or worse sleep quality.

          *Stress Level (Ordinal)* : The primary predictor. We can test if a person's self-reported stress level (on a scale, e.g., 1–10) is associated with lower quality of sleep.

          *BMI Category (Categorical)* : We can test if being in the "Obese" or "Overweight" category is associated with lower sleep quality compared to "Normal".

          *Occupation (Categorical)* : We can test if certain occupations (e.g., high-stress jobs like "Doctor" or "Nurse") are associated with lower sleep quality.

      3.  **Confounding Variables**: A third, "lurking" variable that is associated with both the independent variable and the dependent variable. It can create a false association or mask a real one, making it seem like the independent variable is causing the change in the dependent variable when it might not be the true cause.
     
          Potential Confounding Variables in this Study:
          
          *Age* : Age is a major potential confounder.

          *Why it's confounding* : Age affects stress levels (younger adults might have work or school stress; older adults might have health-related stress). Age also affects sleep quality (sleep patterns change with age, and older adults often report poorer sleep).

          *How to deal with it* : During analysis, we can compare the effect of Age on Quality of Sleep within specific age groups, e.g., 20-30, 30-40, etc. This is also known as Stratified Analysis. This involves:

          i) Divide participants into age groups (strata), for example: 20–30, 31–40, 41–50, etc.

          ii) Analyze the relationship between stress level and sleep quality within each age group separately.

          iii) This helps you see if the relationship holds consistently across different ages.

# Detecting outliers using Z-score method

After looking for potential type inconsistencies, problematic misspellings, In the above step, I am set to identify potential outliers in the numeric attributes of the dataset using the Z-score method.

1). Firstly, I select all numeric columns in the dataset.

2). For each numeric column, calculate the Z-score for each value. The Z-score measures how many standard deviations a value is away from the column mean.

3). Values with an absolute Z-score greater than 3 are considered outliers, as they are unusually far from the typical range of the data.

4). Null values are dropped before calculation to ensure accurate results.

5). Any detected outliers are printed for further inspection, helping me identify extreme or unusual values that may affect analysis.




```python
# Identify numeric columns
numeric_cols = df.select_dtypes(include=[np.number]).columns

# Define known invalid placeholder values
invalid_placeholders = [-999, 9999, 5000, -5000]

# Loop through each numeric column
for col in numeric_cols:
    col_data = df[col]

    # Replace known invalid placeholders
    df.loc[df[col].isin(invalid_placeholders), col] = np.nan

    # Compute Z-scores (excluding NaNs)
    z_scores = np.abs(stats.zscore(df[col].dropna()))

    # Identify statistical outliers (Z > 3)
    outlier_indices = df[col].dropna().index[z_scores > 3]

    # Replace those outliers with NaN
    df.loc[outlier_indices, col] = np.nan

# Confirm cleaning
print("Outlier cleaning complete.")
print(df[numeric_cols].describe())
```

```python
# Update the agg_dict to use the new column names
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = df.select_dtypes(include='object').columns.tolist()

agg_dict = {col: 'mean' for col in numeric_cols if col not in ['Person ID']} # Exclude Person ID from mean calculation
for col in categorical_cols:
    # Join text columns or pick first value for categorical, handling empty mode()
    agg_dict[col] = lambda x: ' | '.join(x.dropna().astype(str).unique()) if x.nunique() > 1 else (x.mode()[0] if not x.mode().empty else np.nan)

# Aggregate duplicates by Person ID
df = df.groupby('Person ID').agg(agg_dict).reset_index()


# Verify aggregation
print(df.head())
print("Duplicates left:", df.duplicated(subset='Person ID').sum())
```

# Cleaning Categorical Columns

***Gender:***

Removed special characters and extra spaces.

Standardized values to Male and Female.

Filled missing values with Unknown.

***Occupation:***

Cleaned text by removing special characters and extra spaces.

Filled missing entries with Unknown.

***BMI Category:***

Cleaned text and removed special characters.

Standardized categories to Normal, Overweight, and Obese.

Filled missing values with Unknown.

***Blood Pressure:***

Removed special characters while keeping numeric values and slashes intact.

Filled missing values with Unknown.

***Sleep Disorder:***

Cleaned text and removed special characters.

Standardized categories to None, Sleep Apnea, and Insomnia.

Filled missing values with None.



Checked unique values in all categorical columns to confirm cleaning and standardization.

```python
pd.set_option('display.max_colwidth', None)
```

```python
import pandas as pd
import re
import numpy as np

# Function to clean text: remove special characters and extra spaces
def clean_text(text):
    if pd.isnull(text):
        return text
    # Remove non-alphanumeric characters except spaces and slash (for blood pressure)
    text = re.sub(r'[^A-Za-z0-9 /]', '', str(text))
    # Remove multiple spaces
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# 1. Clean Gender
df['Gender'] = df['Gender'].apply(clean_text)
df['Gender'] = df['Gender'].replace({'Male': 'Male', 'Female': 'Female'})
df['Gender'] = df['Gender'].fillna('Unknown')

# 2. Clean Occupation
df['Occupation'] = df['Occupation'].apply(clean_text)
df['Occupation'] = df['Occupation'].fillna('Unknown')

# 3. Clean BMI Category
df['BMI Category'] = df['BMI Category'].apply(clean_text)
# Standardize common categories
df['BMI Category'] = df['BMI Category'].replace({
    'Normal Weight': 'Normal',
    'Normal': 'Normal',
    'Overweight': 'Overweight',
    'Obese': 'Obese'
})
df['BMI Category'] = df['BMI Category'].fillna('Unknown')

# 4. Clean Blood Pressure
df['Blood Pressure (mmHg)'] = df['Blood Pressure (mmHg)'].apply(clean_text)
df['Blood Pressure (mmHg)'] = df['Blood Pressure (mmHg)'].fillna('Unknown')

# 5. Clean Sleep Disorder
df['Sleep Disorder'] = df['Sleep Disorder'].apply(clean_text)
df['Sleep Disorder'] = df['Sleep Disorder'].replace({
    'None': 'None',
    'Sleep Apnea': 'Sleep Apnea',
    'Insomnia': 'Insomnia'
})
df['Sleep Disorder'] = df['Sleep Disorder'].fillna('None')

# Check unique values after cleaning
categorical_cols = ['Gender','Occupation','BMI Category','Blood Pressure (mmHg)','Sleep Disorder']
for col in categorical_cols:
    print(f"{col} unique values after cleaning: {df[col].unique()}")
```

Numeric columns with impossible values are being replaced with NaN so that they can later be handled using median or mean imputation.
This avoids skewing the data with extreme incorrect numbers.

```python
# Replace invalid/outlier values with NaN
df['Age'] = df['Age'].replace([5000, 9999], np.nan)
df['Sleep Duration'] = df['Sleep Duration'].replace([-999, 5000], np.nan)
df['Quality of Sleep'] = df['Quality of Sleep'].replace([-999], np.nan)
df['Physical Activity Level'] = df['Physical Activity Level'].replace([9999], np.nan)
df['Stress Level'] = df['Stress Level'].replace([-999], np.nan)
df['Heart Rate (bpm)'] = df['Heart Rate (bpm)'].replace([9999], np.nan)
```

Handling missing values after replacing outliers by replacing NaN with median for numeric columns.

```python
numeric_cols = ['Age','Sleep Duration','Quality of Sleep','Physical Activity Level','Stress Level','Heart Rate (bpm)']
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())
```

Printing the cleaned stats

```python
for col in numeric_cols:
    print(f"{col} stats after cleaning:")
    print(df[col].describe())
    print("------")
```

**2).PCA (Principal Component Analysis)**

*   Looks at how much variance each feature contributes.
*   Creates new, uncorrelated “principal components” that capture most of the information.

**Difference in dimensionality reduction**:

There is a difference between dimensionality reduction performed by both Correlation and PCA. 3. Speeds up computation: Fewer features means fewer computations for distance calculations means faster KNN.

4. Reduces risk of overfitting: Models become simpler and generalize better.

# Assumptions the KNN algorithm makes
KNN does not rely on strong statistical assumptions (e.g., normality), but it implicitly makes several assumptions for good performance:

1. Assumption of Local Similarity: KNN assumes that similar points exist near each other in feature space. In other words, people with similar stress, sleep, and activity patterns likely have similar sleep disorder outcomes.

2. Importance of Feature Scaling: KNN assumes all features contribute equally to distance calculations. That is,

*   Features must be on the same scale
*   Larger-scale variables (like "Daily Steps") cannot dominate smaller ones (“Sleep Duration”)

*   To mitigate this I will apply StandardScaler before fitting KNN futher.
No explicit distributional assumptions

3. Unlike logistic regression or linear regression, KNN does not assume:

*   normality
*   linearity
*   specific data distributions

4. Assumes no extreme outliers: Because KNN uses distance, outliers can distort neighbor relationships. To mitigate this I already removed outliers using Z-score in the beginning

5. **Sample size considerations**: KNN performs best when:

*   The dataset is not extremely large (because prediction time grows with sample size)
*   The dataset is not extremely small (because the neighborhood structure becomes meaningless)

6. Assumes reasonable independence between features:

*   While KNN doesn’t require full independence, highly correlated features can bias the distance metric. To mitigate this problem, PCA is already performed removing highly redundant correlated features.


# Issues with KNN, Why they happen, and how to mitigate them

1). **Overfitting**

Why It Happens:

KNN predicts a class based on the closest neighbors.
When k is too small (for example, k = 1 or k = 2), the model bases its prediction on just one or two data points.
This makes the classifier extremely sensitive to noise, outliers, and very local fluctuations in the data.

Mitigation:

Using k-value tuning, commonly through GridSearchCV, to find a k that balances bias and variance.
Trying a range of k values (e.g., 3–25) helps identify the value that provides the best generalization performance on validation data.

2). **Underfitting**

Why It Happens:

When k is too large, the algorithm considers many neighbors, sometimes even points that are far away and not very similar.
This causes oversmoothing, where the model becomes too simple and cannot properly capture meaningful patterns.

Mitigation:

It can be solved through hyperparameter tuning.
A common heuristic starting point is k= √n, where n is the number of samples.
From there, cross-validation is used to choose the best k that avoids both underfitting and overfitting.

3). **Sensitivity to Scaling** (Specific to my Project)

Why It Happens:

KNN uses distance-based measurements (usually Euclidean distance).
Features with larger numerical ranges (e.g., Daily Steps in the thousands vs. Stress Level on a 1–10 scale) will dominate the distance calculation, even if they are not more important.

Mitigation:

Applying StandardScaler (or MinMaxScaler) so all features are on a comparable scale.
This ensures that no single feature disproportionately influences distance computations.

# Evaluating my Machine Learning Model:

Since my machine learning task is binary classification (predicting whether a person has a sleep disorder: yes = 1, no = 0), the most appropriate evaluation metrics are:

1. Accuracy

   Gives the overall percentage of correctly predicted cases.

   Useful as a first, general indicator of model performance.

2. Confusion Matrix

   Shows true positives, true negatives, false positives, and false negatives.

   Helps understand what kinds of mistakes the model is making.

   Important when false negatives have health implications.

3. Precision

   Of all predicted sleep disorder cases, how many were correct?

   Important if false alarms should be minimized.

4. Recall

   Of all actual sleep disorder cases, how many did we correctly identify?

   Important in health-related problems because missing real cases is costly.

5. F1-Score

   Harmonic mean of precision and recall.

   Good when classes may be imbalanced or when we care about both false positives and false negatives.

   These metrics together give a complete picture of model quality instead of relying only on accuracy.

In my opinion:

If false positives are bad (don’t want to scare healthy people): Precision is more important.

If false negatives are dangerous (don’t want to miss real cases): Recall is more important.

So, for health problems (like Sleep Disorder): Recall is usually more important because missing a real disorder is worse than predicting one incorrectly.

# 1. Logistic Regression

I first tried Logistic Regression because it is one of the simplest and most stable classification algorithms. It works well when the data has a roughly linear relationship between the input features and the target.

I trained the model using the same training/test split as KNN and evaluated it using accuracy, precision, recall, F1-score, and a confusion matrix. Logistic Regression gave me a good baseline for comparison because it tends to generalize well and is less sensitive to noise.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train, y_train)

# Predictions
y_pred_log = log_reg.predict(X_test)

# Accuracy, Precision, Recall, F1-score
log_acc = accuracy_score(y_test, y_pred_log)
log_precision = precision_score(y_test, y_pred_log)
log_recall = recall_score(y_test, y_pred_log)
log_f1 = f1_score(y_test, y_pred_log)

print("===LOGISTIC REGRESSION===")
print("Logistic Regression Accuracy:", log_acc)
print("Logistic Regression Precision:", log_precision)
print("Logistic Regression Recall:", log_recall)
print("Logistic Regression F1 Score:", log_f1)

# Classification Report
print("\nClassification Report (Logistic Regression):")
print(classification_report(y_test, y_pred_log))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred_log)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens')
plt.title("Logistic Regression - Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

# 3. Random Forest Classifier

Finally, I tried a Random Forest model, which is an ensemble of multiple Decision Trees. Random Forests usually perform better than a single tree because they average the predictions of many trees, reducing the chance of overfitting.

I trained the Random Forest with 200 trees and evaluated it using the same performance metrics. Random Forest is generally more robust and tends to give more stable predictions across different subsets of data.

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=200, random_state=42)
rf_model.fit(X_train, y_train)

# Predictions
y_pred_rf = rf_model.predict(X_test)

# Accuracy, Precision, Recall, F1-score
rf_acc = accuracy_score(y_test, y_pred_rf)
rf_precision = precision_score(y_test, y_pred_rf)
rf_recall = recall_score(y_test, y_pred_rf)
rf_f1 = f1_score(y_test, y_pred_rf)

print("===RANDOM FOREST===")
print("Random Forest Accuracy:", rf_acc)
print("Random Forest Precision:", rf_precision)
print("Random Forest Recall:", rf_recall)
print("Random Forest F1 Score:", rf_f1)

# Classification Report
print("\nClassification Report (Random Forest):")
print(classification_report(y_test, y_pred_rf))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred_rf)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Purples')
plt.title("Random Forest - Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
